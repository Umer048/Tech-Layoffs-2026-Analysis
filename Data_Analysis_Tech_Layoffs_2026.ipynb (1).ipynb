```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('fivethirtyeight')
```

**"I started by importing the essential Python libraries: Pandas for data manipulation, NumPy for numerical operations, and Matplotlib/Seaborn for creating professional visualizations. I also applied the 'fivethirtyeight' style to ensure the charts look clean and modern."**

## 1. Data Acquisition

```python
df = pd.read_excel("D:/tech_layoffs_2026_tracker.xlsx")
```

```python
print(df.shape)
```

```python
df.head()
```

**"In this step, I loaded the real-world 2026 tech layoffs dataset. Using .shape and .head(), I performed a quick inspection to understand the structure of the data and verify that all columns were imported correctly."**

## 2. Data Cleaning & Preparation

```python
df['layoff_date'] = pd.to_datetime(df['layoff_date'])
```

```python
print(df.isnull().sum())
```

```python
df['jobs_cut'] = df['jobs_cut'].fillna(0)
```

**"Data cleaning is crucial for accuracy. I converted the date column into a proper datetime format and handled missing values in the jobs_cut column. This ensures that my calculations and time-series charts will be reliable."**

## 3. Exploratory Data Analysis

### 3.1 Top 10 Companies by Jobs Cut

```python
top_layoffs = df.groupby('company')['jobs_cut'].sum().sort_values(ascending=False).head(10)
```

```python
plt.figure(figsize=(10,6))
top_layoffs.plot(kind='bar', color='darkred')
plt.title('Top 10 Companies with Highest Layoffs (Q1 2026)')
plt.ylabel('Number of Jobs Cut')
plt.show()
```

**"I aggregated the total job cuts by company to identify the biggest players in the 2026 layoff wave. The bar chart clearly shows that Mega-scale companies like Oracle and Amazon are leading the workforce reduction this quarter."**

## 4. The AI Influence Factor

### 4.1 Jobs Cut: AI vs Non-AI Reasons

```python
ai_impact = df.groupby('ai_cited')['jobs_cut'].sum()
```

```python
ai_impact.plot(kind='pie', autopct='%1.0f%%', colors=['red', 'blue'])

plt.title("AI vs Non-AI Layoffs") # Title lazmi dein taakay LinkedIn par acha lagay
plt.show()
```

**"This is the core of my analysis. By grouping the data by the ai_cited flag, I visualized the proportion of layoffs driven by AI transformation. It is fascinating to see how many companies are explicitly citing AI as a reason for restructuring."**

## 5. Market Reaction to Layoffs

```python
# Stock reaction vs AI
sns.boxplot(x='ai_cited', y='stock_change_day_pct', data=df)
plt.title('Stock Market Reaction: AI Cited vs Others')
plt.show()
```

**"Finally, I analyzed the stock market's reaction to these announcements. Using a Boxplot, I compared the daily stock change for companies citing AI versus those that did not. This reveals whether investors are rewarding the shift toward AI infrastructure."**
